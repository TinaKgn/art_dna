```python
artists_dataframe = pd.read_csv('artists_database/artists.csv')
artists_dataframe.head()
```

```python
#artists_dataframe.shape
#artists_dataframe.info()
#artists_dataframe.isnull().sum()
```

```python
artists_genres = artists_dataframe['genre'].apply(lambda x: [genres.strip() for genres in x.split(',') if genres])
```

```python
mlb = MultiLabelBinarizer()
genres_encoded = mlb.fit_transform(artists_genres)
genres_dataframe = pd.DataFrame(genres_encoded, columns=mlb.classes_, index=artists_dataframe.index)
```

```python
artists_genre_dataframe = pd.concat([artists_dataframe, genres_dataframe], axis=1)
artists_genre_dataframe.head()
```
